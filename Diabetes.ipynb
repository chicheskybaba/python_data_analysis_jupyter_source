# 12.1 Selecting Best Models Using Exhaustive Search

```python
# import dataset
df=pd.read_csv('diabetes.csv')
df.head(5)
```

```python
df.isna().sum().sum()
```

```python
df.shape
```

```python
features = np.array(df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 
                        'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']])
features
```

```python
target = np.array(df[['Outcome']])
target
```

```python
# Use scikit-learn’s GridSearchCV:
# Load libraries
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import GridSearchCV
```

```python
# Create logistic regression
logistic = linear_model.LogisticRegression(max_iter=100000)
```

```python
# Create range of candidate penalty hyperparameter values
penalty = ['l1', 'l2']
```

```python
# Create range of candidate regularization hyperparameter values
C = np.logspace(0, 4, 10)
```

```python
# Create dictionary hyperparameter candidates
hyperparameters = dict(C=C, penalty=penalty)
```

```python
# Create grid search
gridsearch = GridSearchCV(logistic, hyperparameters, cv=5, verbose=0)
```

```python
# Fit grid search
best_model = gridsearch.fit(features, target)
```

```python
np.logspace(0, 4, 10)
```

```python
# View best hyperparameters
print('Best Penalty:', best_model.best_estimator_.get_params()['penalty'])
print('Best C:', best_model.best_estimator_.get_params()['C'])
```

```python
# Predict target vector
best_model.predict(features)
```

# 12.2 Selecting Best Models Using Randomized Search

```python
# Load libraries
from scipy.stats import uniform
from sklearn import linear_model, datasets
from sklearn.model_selection import RandomizedSearchCV
```

```python
df.head(5)
```

```python
features
```

```python
target
```

```python
# Create logistic regression
logistic = linear_model.LogisticRegression()
```

```python
# Create range of candidate regularization penalty hyperparameter values
penalty = ['l1', 'l2']
```

```python
# Create distribution of candidate regularization hyperparameter values
C = uniform(loc=0, scale=4)
```

```python
# Create hyperparameter options
hyperparameters = dict(C=C, penalty=penalty)
```

```python
# Create randomized search
randomizedsearch = RandomizedSearchCV(
    logistic, hyperparameters, random_state=1, n_iter=100, cv=5, verbose=0,
    n_jobs=-1)
```

```python
# Fit randomized search
best_model = randomizedsearch.fit(features, target)
```

```python
# Define a uniform distribution between 0 and 4, sample 10 values
uniform(loc=0, scale=4).rvs(10)
```

```python
# View best hyperparameters
print('Best Penalty:', best_model.best_estimator_.get_params()['penalty'])
print('Best C:', best_model.best_estimator_.get_params()['C'])
```

```python
# Predict target vector
best_model.predict(features)
```

# 12.3 Selecting Best Models from Multiple Learning Algorithms

```python
# Problem
# You want to select the best model by searching over a range of 
# learning algorithms and their respective hyperparameters.
```

```python
# Solution
# Create a dictionary of candidate learning algorithms and their hyperparameters:
```

```python
# Load libraries
import numpy as np
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
```

```python
# Set random seed
np.random.seed(0)
```

```python
df.head(5)
```

```python
features
```

```python
target
```

```python
# Create a pipeline
pipe = Pipeline([("classifier", RandomForestClassifier())])
```

```python
# Create dictionary with candidate learning algorithms and their hyperparameters
search_space = [{"classifier": [LogisticRegression()],
                 "classifier__penalty": ['l1', 'l2'],
                 "classifier__C": np.logspace(0, 4, 10)},
                {"classifier": [RandomForestClassifier()],
                 "classifier__n_estimators": [10, 100, 1000],
                 "classifier__max_features": [1, 2, 3]}]
```

```python
# Create grid search
gridsearch = GridSearchCV(pipe, search_space, cv=5, verbose=0)
```

```python
# Fit grid search
best_model = gridsearch.fit(features, target)
```

```python
# Python program to generate the prime numbers from 1 to N

number_N = input("Enter the number: ")
integer_N = int(number_N)
prime_number = []

for i in range(2, integer_N+1, 1):  # List of integers to be used as divisors for the input integer N
    check = []
    for j in range(2, i, 1):
        remains = (i % j)
        check.append(remains)
    y = check.count(0)
    if y==0:
        prime_number.append(i)
                      
print("The prime from 1 to " + number_N + " are " + str(prime_number))
```
